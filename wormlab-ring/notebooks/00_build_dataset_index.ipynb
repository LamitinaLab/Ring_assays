```python
from pathlib import Path
import pandas as pd

# Define paths
project_root = Path.cwd().parent 
raw_data_dir = project_root / 'data' / 'raw'
processed_data_dir = project_root / 'data' / 'processed'
output_path = processed_data_dir / 'dataset_index.parquet'

print(f"Searching for data in: {raw_data_dir}")

# Find all files ending with _Position.csv in the raw data directory
position_files = list(raw_data_dir.glob('*/*/*_Position.csv'))

print(f"Found {len(position_files)} Position files.")

# Create a list of dictionaries with dataset information
dataset_list = []
for position_file in position_files:
    dataset_id = position_file.parent.name
    genotype = position_file.parent.parent.name
    file_prefix = f"{dataset_id}"

    dataset_list.append({
        'dataset_id': dataset_id,
        'genotype': genotype,
        'position_path': str(position_file),
        'area_path': str(position_file.parent / f'{file_prefix}_Area.csv'),
        'fit_path': str(position_file.parent / f'{file_prefix}_Fit.csv'),
        'length_path': str(position_file.parent / f'{file_prefix}_Length.csv'),
        'labels_path': str(position_file.parent / f'{file_prefix}_Labels.csv') # Corrected to be a file path
    })

# Create a DataFrame and save it to a Parquet file
if dataset_list:
    dataset_index = pd.DataFrame(dataset_list)
    dataset_index.to_parquet(output_path, index=False)
    print(f"Dataset index saved to {output_path}")
    print("\nFirst 5 entries:")
    print(dataset_index.head())
else:
    print("\nNo datasets found. Please check the following:")
    print(f"1. Your raw data directory exists at: {raw_data_dir}")
    print("2. The structure is 'data/raw/<genotype>/<dataset_id>/<dataset_id>_Position.csv'")
```
